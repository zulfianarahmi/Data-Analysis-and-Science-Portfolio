## Age Distribution Analysis:

Understanding the age distribution helps identify age ranges more vulnerable to heart disease, guiding more effective prevention and early detection campaigns.

```python
#Visualizing the Age Distribution (Analyst)
fig = plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Number of Individuals')
plt.show()
```

## Gender Analysis:
Understanding the gender distribution in the dataset can aid in developing targeted solutions based on population characteristics.

```python
# Analyzing Gender Distribution
gender_counts = data['Gender'].value_counts()
gender_counts
```

```python
# Creating a pie chart for gender distribution
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=['Female (0)', 'Male (1)'], autopct='%1.1f%%', colors=['lightcoral', 'peachpuff'], startangle=90)
plt.title('Distribution of Gender')
plt.show()
```

## Relationship between Resting Blood Pressure and Serum Cholesterol

Understanding the relationship between blood pressure and cholesterol helps determine intervention strategies for managing cardiovascular risk factors.

```python
# Visualizing the Distribution of Resting Blood Pressure
plt.figure(figsize=(10, 6))
sns.histplot(data['RestingBP'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Resting Blood Pressure')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
# Visualizing the Distribution of Serum Cholesterol
plt.figure(figsize=(10, 6))
sns.histplot(data['SerumCholestrol'], bins=20, kde=True, color='lightcoral')
plt.title('Distribution of Serum Cholesterol')
plt.xlabel('Serum Cholesterol')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
# Creating a Scatter Plot between Resting Blood Pressure and Serum Cholesterol
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RestingBP', y='SerumCholestrol', data=data, color='green')
plt.title('Scatter plot between Resting Blood Pressure and Serum Cholesterol')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Serum Cholesterol')
plt.show()
```

```python
# Calculating the Correlation Matrix
correlation_matrix = data[['RestingBP', 'SerumCholestrol']].corr()
```

```python
# Visualizing the Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Blood Pressure and Cholesterol')
plt.show()
```

## Chest Pain Type Analysis

In-depth understanding of chest pain types

```python
## Visualizing the Distribution of Chest Pain Types
plt.figure(figsize=(10, 6))
sns.countplot(x='ChestPain', data=data, palette='viridis')
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
# Count the Proportion of Chest Pain Types
chest_pain_counts = data['ChestPain'].value_counts()

# List of labels corresponding to category numbers
labels = ['Typical Angina (0)', 'Atypical Angina (1)', 'Non-Anginal Pain (2)', 'Asymptomatic (3)']

# Visualizing the pie chart of chest pain type proportions with labels
plt.figure(figsize=(8, 8))
plt.pie(chest_pain_counts, labels=labels, autopct='%1.1f%%', colors=sns.color_palette('viridis'), startangle=90)
plt.title('Proportion of Chest Pain Types')
plt.show()
```

## Resting Electrocardiogram Analysis

Early indications of heart issues.

```python
# Visualizing the Distribution of Resting Electrocardiogram Results
plt.figure(figsize=(10, 6))
sns.countplot(x='RestingrElectro', data=data, palette='viridis')
plt.title('Distribution of Resting Electrocardiogram Results')
plt.xlabel('Resting EKG Result')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
# Count the Proportion of Resting Electrocardiogram Results
resting_ecg_counts = data['RestingrElectro'].value_counts()

# List of labels corresponding to category numbers
labels = ['Normal (0)', 'ST-T Wave Abnormality (1)', 'Left Ventricular Hypertrophy (2)']

# Visualizing the pie chart of resting electrocardiogram result proportions
plt.figure(figsize=(8, 8))
plt.pie(resting_ecg_counts, labels=labels, autopct='%1.1f%%', colors=sns.color_palette('viridis'), startangle=90)
plt.title('Proportion of Resting Electrocardiogram Results')
plt.show()
```

## Maximum Heart Rate Analysis
Knowing the distribution of maximum heart rates helps in determining threshold values that may become crucial predictors in heart disease

```python
#Distribution of maximum heart rates as potential predictors.
plt.figure(figsize=(10, 6))
sns.histplot(data=data['MaxHeartRate'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Maximum Heart Rate')
plt.xlabel('Maximum Heart Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculating Descriptive Statistics for Maximum Heart Rate
mean_heart_rate = data['MaxHeartRate'].mean()
median_heart_rate = data['MaxHeartRate'].median()
std_dev_heart_rate = data['MaxHeartRate'].std()

print(f"Mean Maximum Heart Rate: {mean_heart_rate}")
print(f"Median Maximum Heart Rate: {median_heart_rate}")
print(f"Standard Deviation of Maximum Heart Rate: {std_dev_heart_rate}")
```
